```python
# Grouping
# Group by species
grouped = df.groupby('species')
```

```python
# Perform some aggregations
# Example 1: Calculate the mean of each feature for each species
mean_values = grouped.mean()
print("Mean values:\n", mean_values)
```

```python
# Example 2: Calculate the standard deviation of each feature for each species
std_values = grouped.std()
print("\nStandard Deviation values:\n", std_values)
```

```python
# Example 3: Get the size of each group (number of samples per species)
size_values = grouped.size()
print("\nSize of each group:\n", size_values)
```

```python
# Example 4: Get summary statistics for each group
summary_statistics = grouped.describe()
print("\nSummary statistics:\n", summary_statistics)
```
